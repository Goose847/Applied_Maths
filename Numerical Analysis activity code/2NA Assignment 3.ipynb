Deriving the formula:<br>
$f''(x) = \frac{-f(x+2h) + 16f(x+h) - 30f(x) + 16(x-h) - f(x-2h)}{12h^2}$

We begin by taking the talor expansions of the terms in the numerator:<br>
$16f(x+h) = 16f(x) + 16hf'(x) + \frac{16h^2}{2}f''(x) + O(h^3)$<br>
$16f(x-h) = 16f(x) - 16hf'(x) + \frac{16h^2}{2}f''(x) + O(h^3)$<br>
$-f(x+2h) = -f(x) - 2hf'(x) - \frac{4h^2}{2}f''(x) + O(h^3)$<br>
$-f(x-h) = -f(x) + 2hf'(x) - \frac{4h^2}{2}f''(x) + O(h^3)$<br>

Summing these up, we can see:<br>
$16f(x+h) + 16f(x-h) - f(x+2h) - f(x-h) = 30f(x) + 12h^2f''(x) + O(h^3) $<br>

Therefore, rearraging and ommiting the $O(h^3)$ terms gives us that:

$$f''(x) = \frac{-f(x+2h) + 16f(x+h) - 30f(x) + 16(x-h) - f(x-2h)}{12h^2}$$

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format ='retina' #Increases resolution of plots
```

```python
def f(x):
    return np.exp(x)

def fn_prime(n, x):
    return ( f(x+10**-n) - f(x) ) / ( 10**-n )
```

```python
x = 2
approx = []
h_n = []
for n in range(1, 21, 1):
    approx.append( fn_prime(n, x) )
    h_n.append(10**-n) #for later
    print(approx[n-1])
```

### A
The approximations start with a considerable amount of error which gradually decrease before increasing again.

### B

```python
f_prime = np.exp(2)
abs_err = []
for i in range(0, len(approx), 1):
    abs_err.append( np.absolute( f_prime - approx[i] ) )
lae = np.log(abs_err)
lhn = np.log(h_n)
plt.figure(figsize = (20,10))
plt.gca().invert_xaxis()
plt.loglog(h_n, abs_err, marker = 'o')
plt.grid()
```
