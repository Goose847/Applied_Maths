# Question 1

```python
from scipy.integrate import odeint
import numpy as np
import scipy.integrate as sp
import matplotlib.pyplot as plt
from ipywidgets import *
import scipy.constants as sp
%config InlineBackend.figure_format ='retina'
```

```python

mu = 10

def vanderpol(X, t):
    x = X[0]
    y = X[1]
    dxdt = mu * (x - 1./3.*x**3 - y)
    dydt = x / mu
    return [dxdt, dydt]

X0 = [1, 0]
t = np.linspace(0, 45, 1000)

sol = odeint(vanderpol, X0, t)


x = sol[:, 0]
y = sol[:, 1]

plt.figure(figsize = (10, 8))
plt.grid()
plt.plot(t,x)
plt.title("Solution of the van der Pol Equation")
plt.ylabel(r'$x(t)$')
plt.xlabel(r'$t$')
plt.legend(('x', 'y'))


# phase portrait
plt.figure(figsize = (10, 8))
plt.grid()
plt.plot(x,y)
plt.plot(x[0], y[0], 'ro')
plt.title("Phase Portrait of the van Der Pol equation")
plt.xlabel(r'$x(t)$')
plt.ylabel(r'$y(t)$')
```

# Question 2
## a

```python
def questionTwoPlotter(mu, I, xo, yo, Q): #mu value, the torque, initial x, initial y
    def vanderpol(X, t):
        x = X[0]
        y = X[1]
        dxdt = y
        dydt = I - np.sin(x)-mu*y
        return [dxdt, dydt]

    X0 = [xo, yo]
    t = np.linspace(0, Q, 1000)
    
    sol = odeint(vanderpol, X0, t)
    
    x = sol[:, 0]
    y = sol[:, 1]
    
    solArr = [x, y, t]
    
    return solArr
```

```python
oneAFirst = questionTwoPlotter(0.5, 0.9, 0.1, 0.1, 40)
oneASecond = questionTwoPlotter(0.5, 0.9, 0.2, 0.2, 40)
oneAThird = questionTwoPlotter(0.5, 0.9, 0.3, 0.3, 40)
oneAForth = questionTwoPlotter(0.5, 0.9, 0.4, 0.5, 40)
oneAFifth = questionTwoPlotter(0.5, 0.9, 0.5, 0.6, 40)
```

```python
# phase portrait
plt.figure(figsize = (10, 8))
plt.grid()
plt.plot(oneAFirst[0],oneAFirst[1], label = "(0.1, 0.1)")
plt.plot(0.1, 0.1, 'ro')

plt.plot(oneASecond[0],oneASecond[1], label = "(0.2, 0.2)")
plt.plot(0.2, 0.2, 'ro')

plt.plot(oneAThird[0],oneAThird[1], label = "(0.3, 0.3)")
plt.plot(0.3, 0.3, 'ro')

plt.plot(oneAForth[0],oneAForth[1], label = "(0.4, 0.5)")
plt.plot(0.4, 0.5, 'ro')

plt.plot(oneAFifth[0],oneAFifth[1], label = "(0.5, 0.6)")
plt.plot(0.5, 0.6, 'ro')
plt.legend()
plt.title(r'Phase Portrait of Josephson Junciton for $\mu = 0.5$ and $I = 0.9$')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\dot{\theta}$')
plt.savefig('question2a.png')

print(oneAFirst[0][-1])
print(oneAFirst[1][-1])
```
